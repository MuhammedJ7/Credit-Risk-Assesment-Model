```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
CreditRisk =pd.read_csv('credit_risk_dataset.csv')
CreditRisk.head()
```

```python
CreditRisk.shape
```

```python
CreditRiskCopy=  CreditRisk.copy()
```

```python
cr_age_rmvd = CreditRisk[CreditRisk['person_age']<=70]
cr_age_rmvd.reset_index(drop = True, inplace = True)
cr_age_rmvd.head()
```

```python
person_emp_rmvd= cr_age_rmvd[cr_age_rmvd['person_emp_length']<47]
person_emp_rmvd.reset_index(drop = True, inplace = True)
person_emp_rmvd.head()
```

```python
person_emp_rmvd.isnull().sum()
```

```python
crData=person_emp_rmvd.copy()
crData.shape
```

```python
crData.fillna({'loan_int_rate':crData['loan_int_rate'].median()},inplace = True)
crData.isnull().sum()
```

```python
crData.describe()
```

```python
crDataCopy =crData.drop('loan_grade', axis=1)
display(crDataCopy.shape)
crDataCopy.head()
```

## Categorical Features Treament

```python
crdataCat_tmnt = crDataCopy.copy()
```

```python
person_home_ownership=pd.get_dummies (crdataCat_tmnt['person_home_ownership'],drop_first=True).astype(int)
loan_intent= pd.get_dummies (crdataCat_tmnt['loan_intent'],drop_first=True).astype(int)
crdataCat_tmnt ['cb_person_default_on_file_binary'] = np.where(crdataCat_tmnt['cb_person_default_on_file']=='Y',1,0)
crdataCat_tmnt.head()
```

```python
Datascaling= crdataCat_tmnt.drop(['person_home_ownership','loan_intent','loan_status','cb_person_default_on_file','cb_person_default_on_file_binary'],axis =1)
Datascaling.head()
```

```python
scaler= StandardScaler ()
scaledData = scaler.fit_transform(Datascaling)
scaled_df= pd.DataFrame(scaledData, columns = ['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length'])
scaled_df.head()
```

```python
scaled_data_combined = pd.concat([scaled_df,person_home_ownership,loan_intent],axis = 1)
scaled_data_combined['cb_person_default_on_file'] = crdataCat_tmnt['cb_person_default_on_file_binary']
scaled_data_combined['loan_status'] = crdataCat_tmnt['loan_status']
scaled_data_combined.head()
```

```python
target = scaled_data_combined['loan_status']
features = scaled_data_combined.drop('loan_status', axis = 1 )
features.head()
```

## SMOTE - Synthetic Minority Over-Sampling technique

```python
from imblearn.over_sampling import SMOTE
smote= SMOTE()
balanced_features, balanced_target = smote.fit_resample(features, target)
```

```python
balanced_target.shape# there is an increase because of oversampling of minority class, in our case the loan status of 1 , so it has created synthetic data for that class to balance
```

```python
balanced_target_df = pd.DataFrame ({'target': balanced_target})
balanced_target_df.groupby('target').size()
```

## Models Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split (balanced_features, balanced_target, test_size=0.20, random_state=42)
logit = LogisticRegression()
logit.fit(x_train, y_train)
```

```python
logit.score(x_train,y_train)
```

```python
logit_prediction = logit.predict(x_test)
print(classification_report(y_test, logit_prediction))
print(logit.coef_[0])

features_imp_logit = pd.DataFrame ({'features' :balanced_features.columns,'logit_imp' : logit.coef_[0]})
```
